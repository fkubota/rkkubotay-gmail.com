# hilbert trainsformation

- ヒルベルト変換をためす

# Import everything I need :)

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

# Do it

```python
n = 1000                         # データ数
dt = 0.001                       # サンプリング間隔
t = np.linspace(1, n, n)*dt-dt
y = np.sin(2*np.pi*t)*np.sin(2*20*np.pi*t)

plt.figure(figsize=(20, 3))
plt.plot(t, y)
```

<br>

ヒルベルト変換

```python
y_hil = signal.hilbert(y)
y_hil_real = np.real(y_hil)
y_hil_imag = np.imag(y_hil)
```

```python
plt.figure(figsize=(20, 3))
plt.plot(t, y, label='raw')
plt.plot(t, y_hil_imag, '-.', label='hilbert')
plt.plot(t, np.abs(y_hil), label='envelope', linewidth=5, alpha=0.5)
plt.legend()
```
